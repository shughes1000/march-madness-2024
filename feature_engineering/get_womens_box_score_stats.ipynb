# Get Women's Box Score Stats

Get yearly averages for some stats that you would find in a boxscore

```python
import pandas as pd

pd.set_option('display.max_columns', 100)

include_2024 = True

df = pd.concat(
    [
        pd.read_csv(fr'..\data\unprocessed\womens_sports_reference\womens_full_year_sports_reference_{season}.csv') 
        .assign(Season=season)
        for season in range(2012, 2024 + include_2024)
    ], 
    ignore_index=True
)

df.insert(0, 'Season', df.pop('Season'))

df = df.loc[df['NCAA Tournament'] == 0, :].reset_index(drop=True)

df
```

Fix columns for this analysis

```python
df['Team Win%'] = (df['Result'] == 1).astype(int)

df['Team EFG%'] = (0.5*df['Team 3P'] + df['Team FG'])/df['Team FGA']
df['Opponent EFG%'] = (0.5*df['Opponent 3P'] + df['Opponent FG'])/df['Opponent FGA']

df['Team TOR'] = df['Team TOV'] / df['Possessions'] * 100
df['Opponent TOR'] = df['Opponent TOV'] / df['Possessions'] * 100

df['Team ORBR'] = df['Team ORB'] / df['Possessions'] * 100

df['Team FTR'] = df['Team FT'] / df['Team FGA'] * 100
df['Opponent FTR'] = df['Opponent FT'] / df['Opponent FGA'] * 100

df
```

```python
df_stats = (
    df
    .groupby(['Season', 'Team'])
    [['Team Win%', 'Team EFG%', 'Opponent EFG%', 'Team TOR', 'Opponent TOR', 'Team ORBR', 'Team FTR', 'Opponent FTR']]
    .mean()
    .reset_index()
)

df_stats
```

```python
df_stats.drop(columns=['Opponent TOR'], inplace=True)  # low correlation

df_stats
```

```python
df_stats.to_csv(f'../data/preprocessed/womens_box_score_stats/womens_box_score_stats.csv', index=False)

'Done'
```
