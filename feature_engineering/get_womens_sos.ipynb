# Get Women's Strength of Schedule

Get additional ratings based on strength of opponents

```python
import pandas as pd

pd.set_option('display.max_columns', 100)

include_2024 = True

df = pd.concat(
    [
        pd.read_csv(fr'..\data\unprocessed\womens_sports_reference\womens_full_year_sports_reference_{season}.csv')
        .assign(Season=season)
        for season in range(2012, 2024 + include_2024)
        if season != 2020  # tournament was cancelled
    ],
    ignore_index=True
)

df.insert(0, 'Season', df.pop('Season'))

df = df.loc[df['NCAA Tournament'] == 0, :].reset_index(drop=True)  # only include games before NCAA Tournament

df
```

```python
df_ratings = pd.concat(
    [
        pd.read_csv(fr".\data\preprocessed\womens_my_rankings\womens_my_rankings_{season}.csv")
        .assign(Season=season)
        for season in range(2012, 2024 + include_2024)
        if season != 2020  # tournament was cancelled
    ],
    ignore_index=True
)

df_ratings.insert(0, 'Season', df_ratings.pop('Season'))

df_ratings
```

```python
df = pd.merge(
    df,
    df_ratings[['Season', 'Team', 'Efficiency Margin']].rename(columns={'Team': 'Opponent', 'Efficiency Margin': 'Opponent EM'}),
    how='left',
    on=['Season', 'Opponent'],
)

df
```

```python
df_wins = (
    df
    .loc[df['Result'] == 1, :]
    .groupby(['Season', 'Team', 'Result'])
    ['Opponent EM']
    .quantile([0.50, 0.75, 0.90, 1.00])
    .reset_index()
    .rename(columns={'level_3': 'Quantile'})
    .pivot(index=['Season', 'Team', 'Result'], columns='Quantile', values='Opponent EM')
    .reset_index()
)

df_wins.columns = list(df_wins.columns[:-4]) + list(df_wins.columns[-4:].astype(str) + ' Win')

df_wins
```

```python
df_losses = (
    df
    .loc[df['Result'] == -1, :]
    .groupby(['Season', 'Team', 'Result'])
    ['Opponent EM']
    .quantile([0.00, 0.10, 0.25, 0.50])
    .reset_index()
    .rename(columns={'level_3': 'Quantile'})
    .pivot(index=['Season', 'Team', 'Result'], columns='Quantile', values='Opponent EM')
    .reset_index()
)

df_losses.columns = list(df_losses.columns[:-4]) + list(df_losses.columns[-4:].astype(str) + ' Loss')

df_losses
```

```python
df_sos = pd.merge(
    df_wins.drop(columns=['Result']),
    df_losses.drop(columns=['Result']),
    how='outer',
    on=['Season', 'Team']
)

df_sos
```

There are NAs because some teams did not win or lose. Set these to the EM of the team. 

```python
season_team_to_em = dict(zip(df_ratings['Season'].astype(str) + ' ' + df_ratings['Team'], df_ratings['Efficiency Margin']))

for c in df_sos.columns[2:]:
    df_sos.loc[df_sos[c].isna(), c] = (df_sos.loc[df_sos[c].isna(), 'Season'].astype(str) + ' ' + df_sos.loc[df_sos[c].isna(), 'Team']).map(season_team_to_em)

df_sos
```

```python
df_sos.to_csv('../data/preprocessed/womens_sos/womens_sos.csv', index=False)

'Done'
```
